```python
#import Dependencies
import pandas as pd
import numpy as np
import requests
from pprint import pprint
from sqlalchemy import create_engine
```

## import firstgen pokemon csv

```python
csv_file = "Resources/FirstGenPokemon.csv"
firstgen_df = pd.read_csv(csv_file)
firstgen_df
```

```python
#clean data
#remove spaces from column names
firstgen_df.columns = firstgen_df.columns.str.replace(' ', '')
```

```python
firstgen_cleaned= firstgen_df[['Number','Name','Type1','Height(m)','Weight(kg)']].copy()
firstgen_cleaned.set_index("Number", inplace=True)
firstgen_renamed=firstgen_cleaned.rename(columns={"Number": "number","Name": "name","Type1": "type", "Height(m)": "height_m","Weight(kg)":"weight_kg"})
firstgen_renamed.index.names = ['number']
firstgen_renamed
```

## gather 2nd gen data from API

```python
#create list to hold numbers for 2nd gen pokemon
num_list = [i for i in range(152,252)]

# set up lists to hold reponse info
name=[]
type1=[]
height=[]
weight=[]
number=[]
```

```python
#partial query URL
query_url = "https://pokeapi.co/api/v2/pokemon/"
```

```python
#test out query url
response_1 = requests.get(query_url + '100').json()
pprint(response_1)
```

```python
#query API using loop and save all values to lists
index_counter=1
for num in num_list:
    response = requests.get(query_url + str(num)).json()
    name.append(response['name'])
    height.append(response['height'])
    weight.append(response['weight'])
    number.append(response['id'])
    type1.append(response['types'][0]['type']['name'])
    print(f"Processing Record {index_counter}")
    index_counter += 1
```

```python
# create a data frame from pokemon
secondgen_dict = {
    "number": number,
    "name": name,
    "type": type1,
    "height_m": height,
    "weight_kg": weight
    
}

secondgen_data = pd.DataFrame(secondgen_dict)
secondgen_data =secondgen_data.set_index('number')
secondgen_data.head()
```

## connect to database

```python
connection_string = "postgres:postgres@localhost:5432/Pokemon"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

## load dataframes into database

```python
firstgen_renamed.to_sql(name='firstgen', con=engine, if_exists='append', index=True)
```

```python
secondgen_data.to_sql(name='secondgen', con=engine, if_exists='append', index=True)
```
